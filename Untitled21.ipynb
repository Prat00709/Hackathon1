```python
import tensorflow as tf
from tensorflow.keras import layers, models, losses
from tensorflow.keras.datasets import cifar10
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Load and normalize CIFAR-10
(x_train, _), (x_test, _) = cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Autoencoder architecture
def build_autoencoder():
    input_img = layers.Input(shape=(32, 32, 3))

    # Encoder
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    encoded = layers.MaxPooling2D((2, 2), padding='same')(x)

    # Decoder
    x = layers.Conv2DTranspose(64, (3, 3), strides=2, activation='relu', padding='same')(encoded)
    x = layers.Conv2DTranspose(32, (3, 3), strides=2, activation='relu', padding='same')(x)
    decoded = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

    autoencoder = models.Model(input_img, decoded)
    return autoencoder

autoencoder = build_autoencoder()
autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder
history = autoencoder.fit(
    x_train, x_train,
    epochs=10,
    batch_size=32,
    shuffle=True,
    validation_data=(x_test, x_test)
)
```

```python
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Predict on test data
decoded_imgs = autoencoder.predict(x_test)

# Pick a sample to evaluate
def evaluate_metrics(x_true, x_pred):
    ssim_scores = []
    psnr_scores = []
    mae_scores = []
    mse_scores = []

    for i in range(len(x_true)):
        true_img = x_true[i]
        pred_img = x_pred[i]

        ssim_scores.append(ssim(true_img, pred_img, channel_axis=2, data_range=1.0))
        psnr_scores.append(psnr(true_img, pred_img, data_range=1.0))
        mae_scores.append(mean_absolute_error(true_img.flatten(), pred_img.flatten()))
        mse_scores.append(mean_squared_error(true_img.flatten(), pred_img.flatten()))

    print(f"Avg SSIM: {np.mean(ssim_scores):.4f}")
    print(f"Avg PSNR: {np.mean(psnr_scores):.4f}")
    print(f"Avg MAE: {np.mean(mae_scores):.4f}")
    print(f"Avg MSE: {np.mean(mse_scores):.4f}")

evaluate_metrics(x_test, decoded_imgs)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.datasets import cifar10
import numpy as np

# Load and normalize CIFAR-10
(x_train, _), (x_test, _) = cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

LATENT_DIM = 128

# Encoder
class VAEEncoder(Model):
    def __init__(self, latent_dim):
        super(VAEEncoder, self).__init__()
        self.conv1 = layers.Conv2D(32, 3, activation='relu', padding='same')
        self.pool1 = layers.MaxPooling2D(2, padding='same')
        self.conv2 = layers.Conv2D(64, 3, activation='relu', padding='same')
        self.pool2 = layers.MaxPooling2D(2, padding='same')
        self.flatten = layers.Flatten()
        self.dense_mean = layers.Dense(latent_dim)
        self.dense_log_var = layers.Dense(latent_dim)

    def call(self, x):
        x = self.conv1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = self.pool2(x)
        x = self.flatten(x)
        mean = self.dense_mean(x)
        log_var = self.dense_log_var(x)
        return mean, log_var

# Reparameterization trick
def sample_latent(mean, log_var):
    eps = tf.random.normal(shape=tf.shape(mean))
    return mean + tf.exp(0.5 * log_var) * eps

# Decoder
class VAEDecoder(Model):
    def __init__(self):
        super(VAEDecoder, self).__init__()
        self.dense = layers.Dense(8 * 8 * 64, activation='relu')
        self.reshape_layer = layers.Reshape((8, 8, 64))
        self.deconv1 = layers.Conv2DTranspose(64, 3, strides=2, padding='same', activation='relu')
        self.deconv2 = layers.Conv2DTranspose(32, 3, strides=2, padding='same', activation='relu')
        self.output_layer = layers.Conv2D(3, 3, activation='sigmoid', padding='same')

    def call(self, z):
        x = self.dense(z)
        x = self.reshape_layer(x)
        x = self.deconv1(x)
        x = self.deconv2(x)
        return self.output_layer(x)
```

```python
class VAE(Model):
    def __init__(self, encoder, decoder, beta=1.0):
        super(VAE, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.beta = beta  # weight for KL divergence
        self.total_loss_tracker = tf.keras.metrics.Mean(name="total_loss")
        self.recon_loss_tracker = tf.keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = tf.keras.metrics.Mean(name="kl_loss")

    def train_step(self, data):
        if isinstance(data, tuple):
            data = data[0]
        with tf.GradientTape() as tape:
            data = tf.cast(data, dtype=tf.float32)
            mean, log_var = self.encoder(data)
            z = sample_latent(mean, log_var)
            reconstruction = self.decoder(z)

            recon_loss = tf.reduce_mean(tf.reduce_sum(tf.square(data - reconstruction), axis=[1, 2, 3]))
            kl_loss = -0.5 * tf.reduce_mean(tf.reduce_sum(1 + log_var - tf.square(mean) - tf.exp(log_var), axis=1))
            total_loss = recon_loss + self.beta * kl_loss

        grads = tape.gradient(total_loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.trainable_variables))

        self.total_loss_tracker.update_state(total_loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)

        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.recon_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }
```

```python
encoder = VAEEncoder(latent_dim=LATENT_DIM)
decoder = VAEDecoder()
vae = VAE(encoder, decoder)
vae.compile(optimizer=tf.keras.optimizers.Adam())

EPOCHS = 15
BATCH_SIZE = 64

kl_losses = []
recon_losses = []
total_losses = []

for epoch in range(EPOCHS):
    print(f"Epoch {epoch+1}/{EPOCHS}")
    epoch_kl, epoch_recon, epoch_total = [], [], []

    for i in range(0, len(x_train), BATCH_SIZE):
        x_batch = x_train[i:i+BATCH_SIZE]
        logs = vae.train_step(x_batch)

        epoch_total.append(logs['loss'].numpy())
        epoch_recon.append(logs['reconstruction_loss'].numpy())
        epoch_kl.append(logs['kl_loss'].numpy())

    kl_losses.append(np.mean(epoch_kl))
    recon_losses.append(np.mean(epoch_recon))
    total_losses.append(np.mean(epoch_total))

    print(f" → Loss: {total_losses[-1]:.4f}, Recon: {recon_losses[-1]:.4f}, KL: {kl_losses[-1]:.4f}")
```

```python
# CIFAR-10 labels: 0=airplane, 1=automobile, 2=bird, 3=cat, 4=deer, 5=dog, ...
(_, y_train), (_, y_test) = cifar10.load_data()

def get_sample_by_class(x, y, class_id):
    idx = np.where(y == class_id)[0]
    rand_idx = np.random.choice(idx)
    return x[rand_idx:rand_idx+1]

# Get images
cat_img = get_sample_by_class(x_test, y_test, 3)
dog_img = get_sample_by_class(x_test, y_test, 5)
bird_img = get_sample_by_class(x_test, y_test, 2)

# Encode to latent space
cat_z, _ = encoder(cat_img)
dog_z, _ = encoder(dog_img)
bird_z, _ = encoder(bird_img)

# Latent arithmetic: dog - cat + bird
result_z = dog_z - cat_z + bird_z

# Decode the result
generated_img = decoder(result_z)

# Visualize
fig, axs = plt.subplots(1, 4, figsize=(15, 4))
axs[0].imshow(cat_img[0])
axs[0].set_title("Cat")
axs[1].imshow(dog_img[0])
axs[1].set_title("Dog")
axs[2].imshow(bird_img[0])
axs[2].set_title("Bird")
axs[3].imshow(generated_img[0].numpy())
axs[3].set_title("Result: dog - cat + bird")
for ax in axs:
    ax.axis("off")
plt.suptitle("Latent Space Arithmetic")
plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
import numpy as np
import matplotlib.pyplot as plt

# Load and normalize CIFAR-10
(x_train, _), (x_test, _) = cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
def mask_feature_map(feature_map, mask_ratio=0.25):
    # feature_map shape: (batch, H, W, C)
    batch_size, H, W, C = feature_map.shape
    total_patches = H * W
    num_mask = int(total_patches * mask_ratio)

    masked = tf.identity(feature_map)

    for i in range(batch_size):
        # Get random patch indices
        mask_indices = np.random.choice(total_patches, num_mask, replace=False)
        mask_coords = [(idx // W, idx % W) for idx in mask_indices]

        for (h, w) in mask_coords:
            masked = tf.tensor_scatter_nd_update(
                masked,
                indices=[[i, h, w]],
                updates=[tf.zeros((C,), dtype=feature_map.dtype)]
            )

    return masked
```

```python
class MAEEncoder(tf.keras.Model):
    def __init__(self):
        super(MAEEncoder, self).__init__()
        self.conv1 = layers.Conv2D(32, 3, activation='relu', padding='same')
        self.pool1 = layers.MaxPooling2D(2, padding='same')
        self.conv2 = layers.Conv2D(64, 3, activation='relu', padding='same')
        self.pool2 = layers.MaxPooling2D(2, padding='same')

    def call(self, x):
        x = self.conv1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        return self.pool2(x)  # Output: (8, 8, 64)
```

```python
class MAEDecoder(tf.keras.Model):
    def __init__(self):
        super(MAEDecoder, self).__init__()
        self.deconv1 = layers.Conv2DTranspose(64, 3, strides=2, padding='same', activation='relu')
        self.deconv2 = layers.Conv2DTranspose(32, 3, strides=2, padding='same', activation='relu')
        self.output_layer = layers.Conv2D(3, 3, padding='same', activation='sigmoid')

    def call(self, x):
        x = self.deconv1(x)
        x = self.deconv2(x)
        return self.output_layer(x)
```

```python
class MaskedAutoencoder(tf.keras.Model):
    def __init__(self, encoder, decoder, mask_ratio=0.25):
        super(MaskedAutoencoder, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.mask_ratio = mask_ratio
        self.loss_tracker = tf.keras.metrics.Mean(name="loss")

    def train_step(self, data):
        if isinstance(data, tuple):
            data = data[0]
        with tf.GradientTape() as tape:
            features = self.encoder(data)
            masked_features = mask_feature_map(features, self.mask_ratio)
            recon = self.decoder(masked_features)
            loss = tf.reduce_mean(tf.square(data - recon))  # MSE

        grads = tape.gradient(loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.trainable_variables))
        self.loss_tracker.update_state(loss)
        return {"loss": self.loss_tracker.result()}
```

```python
# Initialize encoder, decoder, and MAE model
mae_encoder = MAEEncoder()
mae_decoder = MAEDecoder()
mae_model = MaskedAutoencoder(mae_encoder, mae_decoder, mask_ratio=0.25)
mae_model.compile(optimizer=tf.keras.optimizers.Adam())

# Train the MAE
mae_history = mae_model.fit(x_train, x_train, epochs=50, batch_size=128, validation_data=(x_test, x_test))
```

```python
# Predict reconstructions from test set
features = mae_encoder(x_test)
masked_features = mask_feature_map(features, mask_ratio=0.25)
mae_reconstructions = mae_decoder(masked_features)

def plot_mae_reconstructions(originals, reconstructions, n=10):
    plt.figure(figsize=(20, 4))
    for i in range(n):
        # Original
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(originals[i])
        plt.title("Original")
        plt.axis("off")

        # Reconstructed
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(reconstructions[i])
        plt.title("MAE Reconstructed")
        plt.axis("off")
    plt.tight_layout()
    plt.show()

plot_mae_reconstructions(x_test, mae_reconstructions.numpy())
```
